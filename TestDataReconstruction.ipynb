```python
import numpy as np
import os
```

```python
#Data is formatted as ASCII tab-separated table with columns for:
#'X','Y','Z','M','R','H','He','C','N','O','Ne','Mg','Si','Fe','epsJ','tform','z','Vr','BE','Lz','pkmassid','acc_label'
#X,Y,Z = 3D position
#M = mass in M_sun
#R = Radial distance from center of galaxy
#H, He, C, N, O, Ne, Mg, Si, Fe = Elemental mass fractions for each particle
#epsJ = particle orbit circularity
#tform = time of formation of the star particle
#z = height above the disc plane
#Vr = radial velocity
#BE = Binding energy
#Lz = Spin angular momentum
#pkmassid = ID of the object this particle used to belong to (if it was accreted)
#acc_label = 0 -> in-situ, 1-> accreted
#build total data
tot_str = []
for f in os.listdir("GalaxyTestData/RAW/TOT_DAT"):
    fi = open("GalaxyTestData/RAW/TOT_DAT/"+f)
    li = fi.readlines()
    #list concatenation is addition in Python!
    #Wooo high level languages
    tot_str += li 
    fi.close()
    print("Incorporated " + f)
#output reconstructed file
ofi = open("GalaxyTestData/Au14.tot_dat","w")
ofi.writelines(tot_str)
ofi.close()
```

```python
#Split the single file into a bunch of smaller files to Github won't choke on it
LINES_PER_FILE = 50000
#Deconstruct the total data file
#GH has a limit of 100Mb per file
#so we'll break the large one down into chunks
file = open("GalaxyTestData/Au14.tot_dat")
lines = file.readlines()
file.close()
N = len(lines)
off = 0
it = 0
curr_str = ""
for l in lines:
    curr_str += l
    it+=1
    if(it >= LINES_PER_FILE or it == N):
        ofile = open("GalaxyTestData/RAW/TOT_DAT/%d.tot_dat"%(off),"w")
        ofile.write(curr_str)
        ofile.close()
        print("Wrote to %d.tot_dat"%(off))
        curr_str = ""
        off += 1
        it = 0
```
